### 請你嘗試觀察每一個「特徵」欄位（除 quality 外）資料的分佈情況，有看出什麼結果嗎？

1. fixed acidity 對於 quality沒有區別性
2. density 對於 quality沒有區別性
3. PH 對於 quality沒有區別性
4. 質量越高的红酒，酒精揮發程度越低
5. 質量越高的紅酒，檸檬酸含量越高

```python
fig, ax = plt.subplots(ncols=5, nrows=3, figsize=(15, 8))
ax = ax.flatten()
index=0
for i in df.columns:
  if i != 'quality':
    sns.barplot(x='quality', y=i, data=df, ax=ax[index])
    index+=1
plt.tight_layout(pad=0.4)
plt.show()
```
